```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from matplotlib.patches import Patch

# Create a 2D histogram for the density plot
plt.figure(figsize=(10, 8))
plt.hist2d(y_true, y_pred, bins=(200, 200), cmap='gray_r', density=True)
plt.colorbar(label='Density')
plt.clim(0, 1)  # Set color limits

# Label the axes
plt.xlabel('Z$_{ELG\ True}$', fontsize=20)
plt.ylabel('Z$_{ELG\ Predicted}$', fontsize=20)

# Set axis limits
plt.xlim(0.8, 1.7)
plt.ylim(0.8, 1.7)

# Perform kernel density estimation
kde = gaussian_kde(np.vstack([y_true, y_pred]))
x_grid = np.linspace(0, 1.8, 200)
y_grid = np.linspace(0, 1.8, 200)
X, Y = np.meshgrid(x_grid, y_grid)
Z_smooth = kde(np.vstack([X.ravel(), Y.ravel()])).reshape(X.shape)

# Define levels for 1, 2, and 3 (percentiles)
levels = [0.68, 0.95, 0.997]  # These correspond to 1, 2, 3 for a normal distribution
contour_levels = [np.percentile(Z_smooth, 100 * level) for level in levels]
contour_levels.append(Z_smooth.max())  # Add an extra level to cover the highest region

# Fill the contours with specified colors
colors = ['orange', 'green', 'purple']
for i, color in enumerate(colors):
    alpha = 0.4 if color != 'purple' else 1  # Set alpha to 1 for blue contour
    plt.contourf(X, Y, Z_smooth, levels=[contour_levels[i], contour_levels[i+1]], colors=[color], alpha=alpha)

# Plot the contour lines
contour_lines = plt.contour(X, Y, Z_smooth, levels=contour_levels, colors='black', linestyles='-', linewidths=1)

# Plot the red dashed line for y = x and add label
plt.plot(np.linspace(0, 1.8, 100), np.linspace(0, 1.8, 100), color='red', linestyle='--', label='y = x')

# Create custom patches for the legend
legend_elements = [
    Patch(facecolor='orange', edgecolor='black', label='1'),
    Patch(facecolor='green', edgecolor='black', label='2'),
    Patch(facecolor='blue', edgecolor='black', label='3')
]

# Add the legend, including the dashed line, in the top right corner
plt.legend(handles=legend_elements + [plt.Line2D([0], [0], color='red', linestyle='--', label='y = x')],
           loc='upper left')

# Save and display the plot
plt.savefig('true_pred_with_sigma_contours_filled.png')
plt.show()
```
